### 4. Thông tin timeline trận đấu

```python
with open(matchID_file_name, 'r', encoding='utf-8') as file:
    matchIDs = json.load(file)
```

```python
matches_Timeline = []
```

```python
def get_match_timeline(api_key, headers, matchIDs):
    base_url = 'https://asia.api.riotgames.com/lol/match/v5/matches/'
    for matchID in matchIDs[:10000]:
        while True:
            url = base_url + matchID + '/timeline?api_key=' + api_key
            response = requests.get(url, headers=headers)
            if response.status_code == 200:
                data = response.json() 
                filtered_data = {
                    "frames": data.get("info", {}).get("frames", []),
                    "participants": data.get("info", {}).get("participants", [])
                }
                matches_Timeline.append(filtered_data)
                matchIDs.remove(matchID)
                break
            elif response.status_code == 429:
                print(len(matches_Timeline))
                sleep(5)
            else:
                print('Wrong')
```

```python
get_match_timeline(api_key, headers, matchIDs)
```

```python
# Ghi dữ liệu
with open(full_match_timeline_file_name, 'w', encoding='utf-8') as file:
    json.dump(matches_Timeline, file, ensure_ascii=False, indent=4)
```
